```python
import = pandas as pd
```

```python
## Load Country Tone Data 
tone_data = pd.read_csv('summary/rekap_all.csv')

# Load Country Tone Data 
WDI_data = pd.read_csv('WDI_Asean.csv')
# tone_data['GDP growth'] = 0


# GDP Growth
data = WDI_data.loc[WDI_data['Indicator Code'] == 'NY.GDP.MKTP.KD.ZG'].sort_values(by=['Country Code']).iloc[:, 48:63].values.ravel()
tone_data['GDP growth'] = data

# Inflation
data = WDI_data.loc[WDI_data['Indicator Code'] == 'NY.GDP.DEFL.KD.ZG'].sort_values(by=['Country Code']).iloc[:, 48:63].values.ravel()
tone_data['Inflation'] = data

# Foreign Investment
data = WDI_data.loc[WDI_data['Indicator Code'] == 'BX.KLT.DINV.WD.GD.ZS'].sort_values(by=['Country Code']).iloc[:, 48:63].values.ravel()
tone_data['FDI'] = data


# CAB
data = WDI_data.loc[WDI_data['Indicator Code'] == 'BN.CAB.XOKA.GD.ZS'].sort_values(by=['Country Code']).iloc[:, 48:63].values.ravel()
tone_data['CAB'] = data

# Unemployment
data = WDI_data.loc[WDI_data['Indicator Code'] == 'SL.UEM.TOTL.NE.ZS'].sort_values(by=['Country Code']).iloc[:, 48:63].values.ravel()
tone_data['Unemployment'] = data

tone_data.to_csv('data_regresi.csv')

tone_data
```
