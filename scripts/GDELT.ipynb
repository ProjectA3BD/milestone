```python
import requests
import lxml.html as lh

gdelt_base_url = 'http://data.gdeltproject.org/events/'

# get the list of all the links on the gdelt file page
page = requests.get(gdelt_base_url+'index.html')
doc = lh.fromstring(page.content)
link_list = doc.xpath("//*/ul/li/a/@href")

# separate out those links that begin with four digits 
file_list = [x for x in link_list if str.isdigit(x[0:4])]

file_list
```

```python
import os.path
import urllib
import zipfile
import glob
import operator

local_path = './gdelt_data/'

fips_country_code1 = 'RP'
fips_country_code2 = 'VM'
fips_country_code3 = 'BM'
fips_country_code4 = 'CB'
fips_country_code5 = 'LA'
# VM, BM, CB, LA, RP, 

outfilecounter = 2043

for compressed_file in file_list[2044:2132]:
    print(compressed_file),
    
    # if we dont have the compressed file stored locally, go get it. Keep trying if necessary.
    while not os.path.isfile(local_path+compressed_file): 
        print('downloading,'),
        urllib.request.urlretrieve(url=gdelt_base_url+compressed_file, 
                           filename=local_path+compressed_file)
        
    # extract the contents of the compressed file to a temporary directory    
    print('extracting,'),
    z = zipfile.ZipFile(file=local_path+compressed_file, mode='r')    
    z.extractall(path=local_path+'tmp/')
    
    # parse each of the csv files in the working directory, 
    print('parsing,'),
    for infile_name in glob.glob(local_path+'tmp/*'):
        outfile_name1 = local_path+'country/'+fips_country_code1+'/'+fips_country_code1+'%04i.tsv'%outfilecounter
        outfile_name2 = local_path+'country/'+fips_country_code2+'/'+fips_country_code2+'%04i.tsv'%outfilecounter
        outfile_name3 = local_path+'country/'+fips_country_code3+'/'+fips_country_code3+'%04i.tsv'%outfilecounter
        outfile_name4 = local_path+'country/'+fips_country_code4+'/'+fips_country_code4+'%04i.tsv'%outfilecounter
        outfile_name5 = local_path+'country/'+fips_country_code5+'/'+fips_country_code5+'%04i.tsv'%outfilecounter
        
        # open the infile and outfile
        with open(infile_name, mode='r') as infile, open(outfile_name1, mode='w') as outfile1, open(outfile_name2, mode='w') as outfile2, open(outfile_name3, mode='w') as outfile3, open(outfile_name4, mode='w') as outfile4, open(outfile_name5, mode='w') as outfile5:
            for line in infile:
                # extract lines with our interest country code
                if fips_country_code1 in operator.itemgetter(51, 37, 44)(line.split('\t')):    
                    outfile1.write(line)
                    
                if fips_country_code2 in operator.itemgetter(51, 37, 44)(line.split('\t')):    
                    outfile2.write(line)
                    
                if fips_country_code3 in operator.itemgetter(51, 37, 44)(line.split('\t')):    
                    outfile3.write(line)
                    
                if fips_country_code4 in operator.itemgetter(51, 37, 44)(line.split('\t')):    
                    outfile4.write(line)
                    
                if fips_country_code5 in operator.itemgetter(51, 37, 44)(line.split('\t')):    
                    outfile5.write(line)
                    
                    
            outfilecounter +=1
            
        # delete the temporary file
        os.remove(infile_name)
        outfile1.close()
        outfile2.close()
        outfile3.close()
        outfile4.close()
        outfile5.close()    
    print('done')
```
