#### c. Struktur Data
Struktur dari data events dari GDELT yang dinilai penting sebagai dasar pengolahan data proyek ini adalah:

| Field | Tipe | Deskripsi
|:---|:---|---:|
| GLOBALEVENTID | Integer | Pengenal unik untuk setiap baris untuk mengidentifikasi data. Catatan: Meskipun akan sering berurutan dengan tanggal, namun tidak selalu terjadi dan bidang ini tidak harus digunakan untuk mengurutkan event berdasarkan tanggal: tanggal seharusnya digunakan. | SQLDATE | String | Tanggal berlangsungnya event dengan format YYYYMMDD. Lihat DATEADDED untuk tanggal YYYYMMDDHHMMSS. | 
| MONTHYEAR | String | Format alternatif tanggal event YYYYMM | 
| Year | String | Format alternatif tanggal event, dalam format YYYY. |
| EventCode | String | berisi kode CAMEO yang menjelaskan tindakan yang dilakukan Actor1 terhadap Actor2.  |
| Actor1Code | String | Kode CAMEO lengkap untuk Actor1 (termasuk geografis, kelas, etnis, agama, dan jenis). Mungkin kosong jika sistem tidak dapat mengidentifikasi Aktor1. |
| Actor2Code | String | Kode CAMEO lengkap untuk Actor2 (termasuk geografis, kelas, etnis, agama, dan tipe-tipe kelas). Mungkin kosong jika Actor2 tidak dapat diidentifikasi. |
| AvgTone | Float | Ini adalah "tone" rata-rata dari semua dokumen yang berisi satu atau lebih mentions dari event ini selama pembaruan 15 menit yang pertama kali dilihat. Skornya berkisar dari -100 (sangat negatif) hingga +100 (sangat positif). Nilai umum berkisar antara -10 dan +10, dengan 0 menunjukkan netral. Field ini dapat digunakan sebagai metode penyaringan "konteks" event sebagai ukuran halus dari pentingnya suatu event dan sebagai proxy untuk "dampak" dari event tersebut itu. Sebagai contoh, sebuah peristiwa kerusuhan dengan tone rata-rata yang sedikit negatif kemungkinan besar merupakan kejadian minor, sedangkan jika memiliki nada rata-rata yang sangat negatif, itu menunjukkan kejadian yang jauh lebih serius. Kerusuhan dengan skor positif kemungkinan menunjukkan kejadian yang sangat kecil yang dijelaskan dalam konteks narasi yang lebih positif (seperti laporan serangan yang terjadi dalam diskusi tentang peningkatan kondisi di lapangan suatu negara dan bagaimana jumlah serangan per hari telah sangat berkurang). |
| GoldsteinScale | Float | Setiap event  code CAMEO diberi skor numerik dari -10 hingga +10, menangkap dampak potensial teoritis yang akan dimiliki oleh jenis event pada stabilitas suatu negara. Ini dikenal sebagai GoldsteinScale. Field ini menentukan score Goldstein untuk setiap jenis event. |
| SOURCEURL | String | Field ini mencatat URL atau kutipan dari laporan berita pertama yang ditemukan dalam suatu event. Dalam sebagian besar kasus ini adalah laporan pertama yang dilihat artikelnya, tetapi karena waktu dan aliran laporan berita melalui jalur pemrosesan, field ini tidak boleh selalu menjadi laporan pertama, tetapi setidaknya dalam beberapa laporan pertama |

#### a. Rekap Event Semua Negara

#### b. Rekap Jumlah Event dan Tone (+/-) Per Tahun
Menampilkan rekapitulasi AvgTone Per Negara berdasarkan tahun yang meliputi Total Event dengan Avg Tone Positif ( > 0), Total Event dengan Avg Tone Negatif ( <0 ), Total Event dengan Avg Netral ( = 0) beserta dengan nilai maksimal dan minimal dari AvgTone pada tahun tersebut.

```python
# Negara: Seluruh Negara

Tone = run("""
SELECT 
    'Malaysia' as 'Negara',
    Year, 
    count(a.GLOBALEVENTID) as `Total Events`,     
    sum(CASE WHEN AvgTone > 0 THEN 1 ELSE 0 END) as `Total Positive Tone`,
    sum(CASE WHEN AvgTone = 0 THEN 1 ELSE 0 END) as `Total Netral Tone`,
    sum(CASE WHEN AvgTone < 0 THEN 1 ELSE 0 END) as `Total Negative Tone`,
    avg(AvgTone) as `Average Tone`,
    max(case when AvgTone >= 0 then AvgTone end) max_positive,
    max(case when AvgTone < 0 then AvgTone end) max_negative
FROM events_MY a GROUP BY Year
UNION
SELECT 
    'Indonesia' as 'Negara',
    Year,     
    count(b.GLOBALEVENTID) as `Total Events`, 
    sum(CASE WHEN AvgTone > 0 THEN 1 ELSE 0 END) as `Total Positive Tone`,
    sum(CASE WHEN AvgTone = 0 THEN 1 ELSE 0 END) as `Total Netral Tone`,
    sum(CASE WHEN AvgTone < 0 THEN 1 ELSE 0 END) as `Total Negative Tone`,
    avg(AvgTone) as `Average Tone`,
    max(case when AvgTone >= 0 then AvgTone end) max_positive,
    max(case when AvgTone < 0 then AvgTone end) max_negative
FROM events_ID b WHERE Year > 2002 GROUP BY Year
UNION
SELECT 
    'Thailand' as 'Negara',
    Year,     
    count(c.GLOBALEVENTID) as `Total Events`, 
    sum(CASE WHEN AvgTone > 0 THEN 1 ELSE 0 END) as `Total Positive Tone`,
    sum(CASE WHEN AvgTone = 0 THEN 1 ELSE 0 END) as `Total Netral Tone`,
    sum(CASE WHEN AvgTone < 0 THEN 1 ELSE 0 END) as `Total Negative Tone`,
    avg(AvgTone) as `Average Tone`,
    max(case when AvgTone >= 0 then AvgTone end) max_positive,
    max(case when AvgTone < 0 then AvgTone end) max_negative
FROM events_TH c GROUP BY Year
UNION
SELECT 
    'Brunei' as 'Negara',
    Year,     
    count(d.GLOBALEVENTID) as `Total Events`, 
    sum(CASE WHEN AvgTone > 0 THEN 1 ELSE 0 END) as `Total Positive Tone`,
    sum(CASE WHEN AvgTone = 0 THEN 1 ELSE 0 END) as `Total Netral Tone`,
    sum(CASE WHEN AvgTone < 0 THEN 1 ELSE 0 END) as `Total Negative Tone`,
    avg(AvgTone) as `Average Tone`,
    max(case when AvgTone >= 0 then AvgTone end) max_positive,
    max(case when AvgTone < 0 then AvgTone end) max_negative
FROM events_BX d GROUP BY Year
UNION
SELECT 
    'Singapore' as 'Negara',
    Year,     
    count(e.GLOBALEVENTID) as `Total Events`, 
    sum(CASE WHEN AvgTone > 0 THEN 1 ELSE 0 END) as `Total Positive Tone`,
    sum(CASE WHEN AvgTone = 0 THEN 1 ELSE 0 END) as `Total Netral Tone`,
    sum(CASE WHEN AvgTone < 0 THEN 1 ELSE 0 END) as `Total Negative Tone`,
    avg(AvgTone) as `Average Tone`,
    max(case when AvgTone >= 0 then AvgTone end) max_positive,
    max(case when AvgTone < 0 then AvgTone end) max_negative
FROM events_SN e GROUP BY Year
  """)
```

```python
# Menyimpan Hasil dan menampilkan Sample

Tone.to_csv('Rekap_AvgTone.csv')
Tone.sample(10)
```

#### c. Rekap Jumlah SOURCE URL per Tahun
Setiap events dalam data GDELT memiliki fielt tautan sumber berita (SOURCEURL). Pada step ini ditampilkan jumlah SOURCEURL yang unik (tidak redundan) dan SOURCEURL yang kosong nilainya.

```python
Source = run("""
SELECT 
    'Malaysia' as 'Negara',
    Year, 
    count(a.GLOBALEVENTID) as `Total Events`, 
    count(DISTINCT a.SOURCEURL) as `Unique Source URL`,
    sum(CASE WHEN SOURCEURL = "n/a" OR SOURCEURL OR SOURCEURL="" IS NULL THEN 1 ELSE 0 END) as `Source URL Kosong`
FROM events_MY a GROUP BY Year
UNION
SELECT 
    'Indonesia' as 'Negara',
    Year, 
    count(b.GLOBALEVENTID) as `Total Events`, 
    count(DISTINCT b.SOURCEURL) as `Unique Source URL`,
    sum(CASE WHEN SOURCEURL = "n/a" OR SOURCEURL OR SOURCEURL="" IS NULL THEN 1 ELSE 0 END) as `Source URL Kosong`
FROM events_ID b Where Year > 2002 GROUP BY Year
UNION
SELECT 
    'Thailand' as 'Negara',
    Year, 
    count(c.GLOBALEVENTID) as `Total Events`, 
    count(DISTINCT c.SOURCEURL) as `Unique Source URL`,
    sum(CASE WHEN SOURCEURL = "n/a" OR SOURCEURL OR SOURCEURL="" IS NULL THEN 1 ELSE 0 END) as `Source URL Kosong`
FROM events_TH c GROUP BY Year
UNION
SELECT 
    'Singapura' as 'Negara',
    Year, 
    count(d.GLOBALEVENTID) as `Total Events`, 
    count(DISTINCT d.SOURCEURL) as `Unique Source URL`,
    sum(CASE WHEN SOURCEURL = "n/a" OR SOURCEURL OR SOURCEURL="" IS NULL THEN 1 ELSE 0 END) as `Source URL Kosong`
FROM events_SN d GROUP BY Year
UNION
SELECT 
    'Brunei' as 'Negara',
    Year, 
    count(e.GLOBALEVENTID) as `Total Events`, 
    count(DISTINCT e.SOURCEURL) as `Unique Source URL`,
    sum(CASE WHEN SOURCEURL = "n/a" OR SOURCEURL IS NULL OR SOURCEURL="" THEN 1 ELSE 0 END) as `Source URL Kosong`
FROM events_BX e GROUP BY Year
  """)
```
